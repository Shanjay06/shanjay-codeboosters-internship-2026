```python
query = """
SELECT department, AVG(math_score) AS average_math_score
FROM students
GROUP BY department
"""
result = pd.read_sql_query(query, conn)
print(result)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(result['department'], result['average_math_score'])
plt.title("Average Math Score by Department")
plt.xlabel("Department")
plt.ylabel("Average Math Score")
plt.xticks(rotation=45)
plt.show()
```

```python
query2 = """
SELECT department, COUNT(*) AS student_count
FROM students
GROUP BY department
"""
panel2 = pd.read_sql_query(query2, conn)
print(panel2)
```

```python
plt.figure(figsize=(7,7))
plt.pie(panel2['student_count'],
        labels=panel2['department'],
        autopct='%1.1f%%')
plt.title("Student Count by Department")
plt.show()
```

```python
query3 = """
SELECT name,
(math_score + science_score + english_score + programming_score) AS total_score
FROM students
ORDER BY total_score DESC
LIMIT 8
"""
panel3 = pd.read_sql_query(query3, conn)
print(panel3)
```

```python
plt.figure(figsize=(10,5))
plt.barh(panel3['name'], panel3['total_score'])
plt.title("Top 8 Students by Total Score")
plt.gca().invert_yaxis()
plt.show()
```

```python
query4 = """
SELECT gender,
ROUND(AVG(attendance_percentage),2) AS avg_attendance
FROM students
GROUP BY gender
"""
panel4 = pd.read_sql_query(query4, conn)
print(panel4)
```

```python
plt.figure(figsize=(6,5))
plt.bar(panel4['gender'], panel4['avg_attendance'])
plt.title("Gender-wise Average Attendance")
plt.show()
```
