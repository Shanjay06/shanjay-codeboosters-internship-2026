```python
import pandas as pd
import sqlite3

df = pd.read_csv('student_performance.csv')

conn = sqlite3.connect('college.db')

df.to_sql('students', conn, if_exists='replace', index=False)

query = """
SELECT department, AVG(math_score)
FROM students
GROUP BY department
"""

result = pd.read_sql_query(query, conn)

print(result)
```
